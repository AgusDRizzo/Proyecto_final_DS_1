```python
#Bucle para calcular el porcentaje de clientes que abandonaron por categoría en cada una de las variables categoricas del dataset. 
#El resultado es una lista, y cada elemento de la lista es un datadrame que contiene como índice las categorías de una variable y como columna CLIENTNUM el porcentaje de clientes que abandonaron
lista_columnas = df.columns
df_attrited = df[df["Attrition_Flag"]=="Attrited Customer"]
lista_df_attrited_division = []
for i in lista_columnas:
    if df.dtypes[i] == "object":
        df_attrited_agrupado = df_attrited.groupby(i)["CLIENTNUM"].count()
        df_t = df.groupby(i)["CLIENTNUM"].count()
        df_attrited_division = (df_attrited_agrupado / df_t)*100
        df_attrited_division = df_attrited_division.to_frame()
        lista_df_attrited_division.append(df_attrited_division)
    else:
        continue
#Probar con el método reset_index()
#Defino variables para establecer el orden de las sitintas categorías en los gráficos de barra
salarios = df["Income_Category"].unique()
tarjetas = df["Card_Category"].unique()
estado_civil = df["Marital_Status"].unique()
eduacion = df["Education_Level"].unique()
salarios = [salarios[1], salarios[3], salarios[0], salarios[2], salarios[4], salarios[5]] 
tarjetas = [tarjetas[0], tarjetas[2], tarjetas[1], tarjetas[3]]
estado_civil = [estado_civil[1], estado_civil[0], estado_civil[3], estado_civil[2]]
eduacion = [eduacion[2], eduacion[0], eduacion[4], eduacion[1], eduacion[5], eduacion[6], eduacion[3]]

#Creo los gráficos de barras usando los dataframes guardados en la list que resultó del bucle for.
f, axs = plt.subplots(2, 3, figsize=(30, 15), gridspec_kw=dict(width_ratios=[5, 5, 5]))
bp0 = sns.barplot(data=lista_df_attrited_division[5], x=lista_df_attrited_division[5].index, y="CLIENTNUM", palette="YlOrBr", ax=axs[0, 0], order = tarjetas)
bp1 = sns.barplot(data=lista_df_attrited_division[4], x=lista_df_attrited_division[4].index, y="CLIENTNUM", palette="YlOrBr", ax=axs[0, 1], order = salarios)
bp2 = sns.barplot(data=lista_df_attrited_division[3], x=lista_df_attrited_division[3].index, y="CLIENTNUM", palette="YlOrBr", ax=axs[0, 2], order = estado_civil)
bp3 = sns.barplot(data=lista_df_attrited_division[2], x=lista_df_attrited_division[2].index, y="CLIENTNUM", palette="YlOrBr", ax=axs[1, 0], order = eduacion)
bp4 = sns.barplot(data=lista_df_attrited_division[1], x=lista_df_attrited_division[1].index, y="CLIENTNUM", palette="YlOrBr", ax=axs[1, 1])

#Cambio el nombre del eje X y del eje Y
bp0.set(xlabel= "Tipo de tarjeta", ylabel= "Porcentaje de clientes que abandonaron")
bp1.set(xlabel= "Categoría de salario", ylabel= "Porcentaje de clientes que abandonaron")
bp2.set(xlabel= "Estado civil", ylabel= "Porcentaje de clientes que abandonaron")
bp3.set(xlabel= "Nivel de educación", ylabel= "Porcentaje de clientes que abandonaron")
bp4.set(xlabel= "Genero", ylabel= "Porcentaje de clientes que abandonaron")


```

```python
#Cálculo del total de clientes agrupados por las distintas variables categóricas
lista_columnas = df.columns
lista_df_totales = []
for i in lista_columnas:
    if df.dtypes[i] == "object":
        df_total_agrupado = df.groupby(i)["CLIENTNUM"].count()
        df_total_agrupado = df_total_agrupado.to_frame()
        lista_df_totales.append(df_total_agrupado)
    else:
        continue

f, axs = plt.subplots(2, 3, figsize=(30, 15), gridspec_kw=dict(width_ratios=[5, 5, 5]))
bp0 = sns.barplot(data=lista_df_totales[5], x=lista_df_totales[5].index, y="CLIENTNUM", palette="YlOrBr", ax=axs[0, 0], order = tarjetas)
bp1 = sns.barplot(data=lista_df_totales[4], x=lista_df_totales[4].index, y="CLIENTNUM", palette="YlOrBr", ax=axs[0, 1], order = salarios)
bp2 = sns.barplot(data=lista_df_totales[3], x=lista_df_totales[3].index, y="CLIENTNUM", palette="YlOrBr", ax=axs[0, 2], order = estado_civil)
bp3 = sns.barplot(data=lista_df_totales[2], x=lista_df_totales[2].index, y="CLIENTNUM", palette="YlOrBr", ax=axs[1, 0], order = eduacion)
bp4 = sns.barplot(data=lista_df_totales[1], x=lista_df_totales[1].index, y="CLIENTNUM", palette="YlOrBr", ax=axs[1, 1])

#Cambio el nombre del eje X y del eje Y
bp0.set(xlabel= "Tipo de tarjeta", ylabel= "Porcentaje de clientes que abandonaron")
bp1.set(xlabel= "Categoría de salario", ylabel= "Porcentaje de clientes que abandonaron")
bp2.set(xlabel= "Estado civil", ylabel= "Porcentaje de clientes que abandonaron")
bp3.set(xlabel= "Nivel de educación", ylabel= "Porcentaje de clientes que abandonaron")
bp4.set(xlabel= "Genero", ylabel= "Porcentaje de clientes que abandonaron")
```
